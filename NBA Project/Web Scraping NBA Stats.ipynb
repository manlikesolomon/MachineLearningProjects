## Web Scraping NBA Stats

```python
# import our libraries
import pandas as pd
import requests as re
from bs4 import BeautifulSoup
import time
```

```python
# store the years we want to scrape for in a list
years = list(range(1991, 2023))
```

```python
# store the url in a variable
url_start = 'https://www.basketball-reference.com/awards/awards_{}.html'
```

```python
# download the html from the web pages and write to a file
for year in years:
    url = url_start.format(year)
    data = re.get(url)
    print(f"{year} - {data.status_code}")
    
    with open('mvp/{}.html'.format(year), 'w+',encoding="utf-8") as file:
        file.write(data.text)
        time.sleep(1)
```

```python
# attempt to scrape the data for a single year before scaling up
with open('data/mvp/1991.html', 'r',encoding="utf-8") as f:
    page = f.read()
```

```python
soup = BeautifulSoup(page)
```

```python
soup.find('tr', class_='over_header').decompose()
```

```python
# parse the table using beautiful soup
mvp_table = soup.find(id='mvp')
```

```python
# read the table with pandas
mvp_1991 = pd.read_html(str(mvp_table))[0]
```

```python
mvp_1991.head()
```

```python
# writing a loop to run this for all our data
dfs = []
i = 0
for year in years:
    with open('data/mvp/{}.html'.format(year), 'r',encoding="utf-8") as file:
        data = file.read()
    soup = BeautifulSoup(data)
    soup.find('tr', class_='over_header').decompose()
    mvp_table = soup.find(id='mvp')
    mvp = pd.read_html(str(mvp_table))[0]
    mvp['year'] = year
    dfs.append(mvp)
```

```python
# combine the dataframes in the list
mvps = pd.concat(dfs)
```

```python
# view the combined dataframe
mvps.head()
```

```python
#save the data in a csv
mvps.to_csv('data/mvps.csv')
```

```python
# scrape all the stats data from our different html files
stats_list = []
for year in years:
    with open('player/{}.html'.format(year), 'r', encoding='utf-8') as file:
        data = file.read()
    soup = BeautifulSoup(data)
    soup.find('tr', class_='thead').decompose()
    stats_table = soup.find(id='per_game_stats')
    stats_df = pd.read_html(str(stats_table))[0]
    stats_df['year'] = year
    stats_list.append(stats_df)
```

```python
# combine the data with pandas 
players_df = pd.concat(stats_list)
players_df.columns = [col.lower() for col in players_df.columns]
```

```python
# save player_stats as csv
players_df.to_csv('data/player_stats.csv')
```

## Scraping Team Stats

```python
# store the team stats url in a variable
team_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

```python
# attempt scraping for one year
year = 1991
url = team_url.format(year)

data = re.get(url)
data.status_code
```

```python
# scrape the html for all years and store in a .html file
for year in years:
    url = team_url.format(year)
    data = re.get(url)
    with open(f'team/{year}.html', 'w+', encoding='utf-8') as file:
        file.write(data.text)
    print(f"{year}-{data.status_code}")
    time.sleep(2)
```

```python
# scrape the data from a single html file before scaling up
year = 1991
with open('team/{}.html'.format(year), 'r', encoding='utf-8') as file:
    page = file.read()
```

```python
# scrape eastern confrence stats
soup = BeautifulSoup(page)
soup.find('tr', class_='thead').decompose()
team_stats_E = soup.find(id='divs_standings_E')
team_E_table = pd.read_html(str(team_stats_E))[0]
team_E_table['confrence'] = 'Eastern'
team_E_table['year'] = year
team_E_table.rename(columns={'Eastern Conference':'team'},inplace=True)
team_E_table.head()
```

```python
# scrape western confrence stats
team_stats_W = soup.find(id='divs_standings_W')
team_W_table = pd.read_html(str(team_stats_W))[0]
team_W_table['confrence'] = 'Western'
team_W_table['year'] = year
team_W_table.rename(columns={'Western Conference':'team'},inplace=True)
team_W_table.head()
```

```python
# scrape all the team stats data using a loop
team_dfs = []
for year in years:
    with open(f'team/{year}.html','r', encoding='utf-8') as file:
        page = file.read()
    soup = BeautifulSoup(page)
    soup.find('tr',class_='thead').decompose()
    # scrape eastern confrence table
    team_stats = soup.find(id='divs_standings_E')
    team_df = pd.read_html(str(team_stats))[0]
    team_df['year'] = year
    team_df['confrence'] = 'Eastern'
    team_df.rename(columns={'Eastern Conference':'team'},inplace=True)
    team_dfs.append(team_df)
    
    # scrape western confrence table
    team_stats = soup.find(id='divs_standings_W')
    team_df = pd.read_html(str(team_stats))[0]
    team_df['year'] = year
    team_df['confrence'] = 'Western'
    team_df.rename(columns={'Western Conference':'team'},inplace=True)
    team_dfs.append(team_df)
```

```python
# concats team stats data into a single dateframe
team_stats_df = pd.concat(team_dfs)
team_stats_df.reset_index(inplace=True)
team_stats_df.drop('index',axis=1,inplace=True)
team_stats_df.sample(10)
```

```python
# save team stats as csv
team_stats_df.to_csv('data/team_stats.csv')
```
